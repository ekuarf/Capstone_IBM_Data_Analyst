## Demo: How to list all tables


```python
# print all the tables names in the database
QUERY = """
SELECT name as Table_Name FROM
sqlite_master WHERE
type = 'table'
"""
# the read_sql_query runs the sql query and returns the data as a dataframe
pd.read_sql_query(QUERY,conn)
```

## Demo: How to run a group by query


```python
QUERY = """
SELECT Age,COUNT(*) as count
FROM master
group by age
order by age
"""
pd.read_sql_query(QUERY,conn)
```

```python
table_name = 'master'  # the table you wish to describe

QUERY = """
SELECT sql FROM sqlite_master
WHERE name= '{}'
""".format(table_name)

df = pd.read_sql_query(QUERY,conn)
print(df.iat[0,0])
```

# Hands-on Lab


## Visualizing distribution of data


### Histograms


Plot a histogram of `ConvertedComp.`


```python
# your code goes here
QUERY = """
SELECT ConvertedComp
FROM master
group by ConvertedComp
"""
ConvertedComp=pd.read_sql_query(QUERY,conn)
print(ConvertedComp.head())

import matplotlib.pyplot as plt
ConvertedComp['ConvertedComp'].plot(kind='hist', bins=20)
plt.title('Histogram of ConvertedComp')
plt.xlabel('Annual Salary (USD)')
plt.ylabel('Frequency')
plt.show()
```

### Box Plots


Plot a box plot of `Age.`


```python
# your code goes here
QUERY = """
SELECT Age
FROM master
"""
Age=pd.read_sql_query(QUERY,conn)
print(Age.head())

Age['Age'].plot(kind='box')
plt.title('Boxplot of the Age Distribution')
#plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
# your code goes here
QUERY = """
SELECT Age, WorkWeekHrs 
FROM master
"""
Age_WWH=pd.read_sql_query(QUERY,conn)
print(Age_WWH.head())

Age_WWH.plot(x='Age', y='WorkWeekHrs', kind='scatter')
plt.title('Age vs. WorkWeekHrs')
plt.xlabel('Age')
plt.ylabel('WorkWeekHrs')
plt.show()
```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
# your code goes here
QUERY = """
SELECT Age, WorkWeekHrs,CodeRevHrs
FROM master
"""
Age_WWH_CRH=pd.read_sql_query(QUERY,conn)
print(Age_WWH_CRH.head())

Age_WWH_CRH.plot(x='WorkWeekHrs', y='CodeRevHrs', kind='scatter', color= 'red', alpha=0.5, s=Age_WWH_CRH['Age']*10+3)
plt.title('Bubbleplot')
plt.xlabel('WorkWeekHrs')
plt.ylabel('CodeRevHrs')
plt.legend(['Age'], loc= 'upper right', fontsize='x-large')
plt.show()
```

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
# SQL Query

QUERY = """
SELECT DatabaseDesireNextYear, COUNT(*) AS Count 
FROM DatabaseDesireNextYear
GROUP BY DatabaseDesireNextYear
ORDER BY Count DESC
LIMIT 5
"""
# the read_sql_query runs the sql query and returns the data as a dataframe
DDNY=pd.read_sql_query(QUERY,conn)
print(DDNY)

#Pie Chart
DDNY['Count'].plot(kind='pie', figsize=(15,6), autopct='%1.1f%%', startangle= 0, labels=None)
plt.title('Top 5 databases to learn next year')
plt.legend(labels=DDNY['DatabaseDesireNextYear'], loc='upper right')
plt.show()
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
# SQL 
QUERY = """
SELECT Age, WorkWeekHrs, CodeRevHrs
FROM master
WHERE Age BETWEEN 30 AND 35
"""
df30=pd.read_sql_query(QUERY,conn)
#print(df30.head())

df_median= df30.groupby('Age').median()
df_median.head()

df_median.plot(kind='bar', stacked=True)
plt.title('Median WorkWeekHrs and CodeRevHrs in the 30-35 age group')
plt.xlabel('Age')
plt.ylabel('Hours')
plt.show()
```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
# SQL
QUERY = """
SELECT Age, ConvertedComp
FROM master
WHERE Age BETWEEN 45 AND 60
"""
df_CC45=pd.read_sql_query(QUERY,conn)

#Calculated median
df_medianCC45= df_CC45.groupby('Age').median()
#df_medianCC45.head()

#plot data
df_medianCC45.plot(kind='line')
plt.title('Median Annual Salary in the 45-60 age group')
plt.xlabel('Age')
plt.ylabel('Annual Salary (USD)')
plt.show()
```

### Bar Chart


Create a horizontal bar chart using column `MainBranch.`


```python
# SQL
QUERY = """
SELECT MainBranch, Count(*) AS Count
FROM master
GROUP BY MainBranch
"""
df_MB=pd.read_sql_query(QUERY,conn)
df_MB.set_index('MainBranch', inplace=True)
print(df_MB.head())

#plot data 
df_MB.plot(kind='barh')
plt.title('Horizontal Bar Chart of MainBranch')
plt.xlabel('Count')

plt.show()
```
